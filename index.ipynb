```python
import pandas as pd

df = pd.read_excel('https://github.com/AxelDelgado1711/practica_02/blob/main/HDI.xlsx?raw=true')
df
```

```python
#Se elimina las 3 primeras filas ya que no sirven
df_cleaned = df.iloc[3:]
df_cleaned
```

```python
#Se eliminará las columnas que tienen nombres NaN
cols_to_keep = df_cleaned.columns[:2] #Se conservará las 2 primeras columnas
df_final = pd.concat([df_cleaned[cols_to_keep], df_cleaned.loc[:, df_cleaned.iloc[0].notna()].iloc[:, 2:]], axis=1)

df_final
```

```python
#Se eliminará todas las filas que sus valores sean NaN, excepto en la segunda columna
df_filtrado = df_final[~(df_final.iloc[:, 2:].isna().all(axis=1) & df_final.iloc[:, 1].notna())]

df_filtrado
```

```python
#Se eliminará las filas que solo tengan valores de NaN
df_filtrado = df_filtrado.dropna(how='all')
df_filtrado
```

```python
#Se va a reemplazar las celdas que tengas valores de NaN con espacios en blanco
df_filtrado = df_filtrado.fillna('')

df_filtrado
```

```python
#Se reemplaza las celdas que tengan ".." y "-" con espacios en blanco
df_filtrado_final = df_filtrado.replace({'..': '', '—': ''})
df_filtrado_final
```

```python
import pandas as pd

df = pd.read_excel('https://github.com/AxelDelgado1711/practica_02/blob/main/HDI.xlsx?raw=true')
df
```

```python
#Se elimina las 3 primeras filas ya que no sirven
df_cleaned = df.iloc[3:]
df_cleaned
```

```python
df_final = df_cleaned.drop(df_cleaned.columns[[3, 5, 7, 9, 11, 13]], axis=1)

df_final
```

```python
#Se eliminará todas las filas que sus valores sean NaN, excepto en la segunda columna
df_filtrado = df_final[~(df_final.iloc[:, 2:].isna().all(axis=1) & df_final.iloc[:, 1].notna())]

df_filtrado
```

```python
#Se eliminará las filas que solo tengan valores de NaN
df_filtrado = df_filtrado.dropna(how='all')
df_filtrado
```

```python
df_filtrado = df_filtrado.dropna(subset=[df_filtrado.columns[0]], axis=0)
df_filtrado.reset_index(drop=True, inplace=True)
df_filtrado
```

```python
df_filtrado.info()
```

```python
df_filtrado[df_filtrado.columns[6]] = df_filtrado[df_filtrado.columns[6]]/1000

df_filtrado
```

```python

for i in range(2, 9):  # De la columna 3 a la 9 (índices 2 a 8)
    df_filtrado.columns.values[i] = f"{df_filtrado.columns[i]}_{df_filtrado.iloc[0, i]}" if pd.notna(df_filtrado.iloc[0, i]) else df_filtrado.columns[i]


df_filtrado = df_filtrado.drop(0)

df_filtrado
```

```python
df_filtrado.iloc[:, 3:7] = df_filtrado.iloc[:, 3:7].round(3)

df_filtrado
```
